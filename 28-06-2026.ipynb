```python
# Ask the user to enter a number
user_input = input("Enter a number: ")
number = int(user_input)

# If the number is even and greater than 50
if number % 2 == 0 and number > 50:
    print("Even and Big")

# If the number is even and less than or equal to 50
elif number % 2 == 0 and number <= 50:
    print("Even and Small")

# If the number is odd and greater than 50
elif number % 1 == 0 and number > 50:
    print("Odd and Big")

# Otherwise (which covers odd and less than or equal to 50)
else:
    print("Odd and Small")
```
