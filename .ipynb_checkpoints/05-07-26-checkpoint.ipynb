```python
cities = ['karachi','Islamabad','Lahore','Quetta',"Hyderabad"]
for city in cities:
    print(city)
```

```python
for u, city in enumerate(cities):
    print(u,city)
```

```python
for index, city in enumerate(cities,1):
    print(index,city)
```

```python
for index, city in enumerate(cities,start=1):
    print(index,city)
```

```python
for x, city in enumerate(cities,start=1):
    print(x,city)
```

```python
#Nested Loap
```

```python
cities = ['karachi','Islamabad','Lahore','Quetta',"Hyderabad"] #city outer loap and visited cities inner loap
visited_cities = ['Islamabad',"Hyderabad","karachi","lahore"]
```

```python
for city in cities:
    for visit in visited_cities:
        print(city,visit)
    print(" ")
```

```python
for city in cities:
    print(city)
    for visit in visited_cities:
        print(visit)
    print(" ")
```

```python
week_days = ['Monday','Tuesday','Wednesday','Thursday','Friday']
time_slots = ['9AM-10AM', '10AM-11AM','11AM-12PM']
for day in week_days:
    print(day)
    for time in time_slots:
        print(time)
    print('-----')
```

```python
for day in week_days:
    for time in time_slots:
        print(day,time)
    print('-----')
```

```python
employee = {
    101 : {'name': 'khan', 'age' : 22, 'Salary' : 2222222},
    102 : {'name': 'alice', 'age' : 33, 'Salary' : 2222},
    103 : {'name': 'john cena', 'age' : 11, 'Salary' : 22}
}
```

```python
employee[104] = {'name': 'sana', 'age': 28, 'Salary': 55000}
```

```python
for emp_id, info in employee.items():
    print(f"Employee ID: {emp_id}")
    print(f"  Name: {info['name']}")
    print(f"  Age: {info['age']}")
    print(f"  Salary: {info['Salary']}")
    print("-" * 20)
```

```python
cities = []
count = ''
while count != 'quit':
    city = input('enter a city')
    if city == 'quit':
        count = 'quit'
    else:
        cities.append(city)
```

```python
#1. Countdown from random number
num = int(input("Enter number: "))

for i in range(num, 0, -1):
    print(i)
```

```python

sum = 0
for i in range(1, 11):
    sum += i
print("Sum =", sum)
```

```python
# 3. Count vowels in a string
string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0

for i in string:
    if i in vowels:
        count += 1

print("Number of vowels:", count)
```

```python
#4. Multiplication table using while loop
# Print multiplication table using while loop
num = int(input("Enter a number: "))
i = 1

while i <= 10:
    print(f"{num} x {i} = {num * i}")
    i += 1
```
