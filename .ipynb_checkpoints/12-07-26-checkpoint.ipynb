## List Comprehension

```python
Ec_Store_Orders = [
    {"order_id": 1, "customer_name": "Ali", "total_amount": 120, "status": "Completed"},
    {"order_id": 2, "customer_name": "Raheel", "total_amount": 80, "status": "Pending"},
    {"order_id": 3, "customer_name": "Saleem", "total_amount": 150, "status": "Completed"},
    {"order_id": 4, "customer_name": "Khushnoor", "total_amount": 95, "status": "Completed"},
    {"order_id": 5, "customer_name": "Alizzy", "total_amount": 200, "status": "Pending"},
    {"order_id": 6, "customer_name": "Javaid", "total_amount": 85, "status": "Completed"},
    {"order_id": 7, "customer_name": "Talha", "total_amount": 150, "status": "Pending"},
    {"order_id": 8, "customer_name": "Naveed", "total_amount": 200, "status": "Completed"},
]
```

```python
for x in Ec_Store_Orders:
    print(x)
```

```python
order_list = []

for order in Ec_Store_Orders:
    if order['total_amount']>100 and order['status'] == 'Completed':
        order_list.append(order['order_id'])
```

```python
order_list
```

The list comprehension iterates through each order dictionary in Ec_Store_Orders

It filters orders where total_amount is greater than 100 AND status is "Completed"

For each matching order, it extracts and includes the order_id in the new list

The result contains order IDs: 1, 3, and 8

```python
completed_high_value_orders = [
    order["order_id"] 
    for order in Ec_Store_Orders 
    if order["total_amount"] > 100 and order["status"] == "Completed"
]
```

```python
print("Order IDs with total_amount > 100 and status 'Completed':")
print(completed_high_value_orders)
```

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A+"
    elif marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "F"

# Test
grade = calculate_grade(85)
print(f"Grade: {grade}")  # Output: Grade: A
```

```python
def simple_interest(principal, rate, time):
    interest = (principal * rate * time) / 100
    total_amount = principal + interest
    return interest, total_amount  # Returns multiple values

# Test
interest, total = simple_interest(10000, 10, 2)
print(f"Interest: Rs. {interest}")  # Output: Interest: Rs. 2000.0
print(f"Total: Rs. {total}")        # Output: Total: Rs. 12000.0
```

```python
def sum_all(*numbers):
    """Add any number of numbers"""
    total = sum(numbers)
    return total

# Test with different numbers of arguments
print(sum_all(1, 2))              # Output: 3
print(sum_all(1, 2, 3))           # Output: 6
print(sum_all(10, 20, 30, 40))    # Output: 100
print(sum_all())                  # Output: 0
```

```python
def average(*numbers):
    """Calculate average of any numbers"""
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)

# Test
print(average(10, 20, 30))        # Output: 20.0
print(average(5, 10, 15, 20, 25)) # Output: 15.0
print(average(100))               # Output: 100.0
print(average())                  # Output: 0
```

```python
def multiply_all(*numbers):
    """Multiply any numbers together"""
    result = 1
    for num in numbers:
        result *= num
    return result

# Test
print(multiply_all(2, 3))          # Output: 6
print(multiply_all(2, 3, 4))       # Output: 24
print(multiply_all(2, 3, 4, 5))    # Output: 120
```

```python
def find_max(*numbers):
    """Find maximum from any numbers"""
    if len(numbers) == 0:
        return None
    return max(numbers)

# Test
print(find_max(10, 25, 5, 30, 15))  # Output: 30
print(find_max(-5, -10, -2))        # Output: -2
```

```python
def combine_strings(*strings):
    """Combine any number of strings"""
    return " ".join(strings)

# Test
print(combine_strings("Hello", "World"))                  # Output: Hello World
print(combine_strings("Python", "is", "awesome"))        # Output: Python is awesome
print(combine_strings("I", "love", "programming"))       # Output: I love programming
```

```python
def calculate_salary(basic_salary, *allowance, **deduction):
    total_allowance = sum(allowance)                    # ✅ Fixed spelling
    total_deduction = sum(deduction.values())           # ✅ Use .values() for dict
    gross_salary = basic_salary + total_allowance
    net_salary = gross_salary - total_deduction         # ✅ No space in variable name
    
    # Display results
    print("=" * 40)
    print("SALARY BREAKDOWN")
    print("=" * 40)
    print(f"Basic Salary:     Rs. {basic_salary}")
    print(f"Total Allowance:  Rs. {total_allowance}")
    print(f"Total Deduction:  Rs. {total_deduction}")
    print("-" * 40)
    print(f"Net Salary:       Rs. {net_salary}")
    print("=" * 40)
    
    return net_salary


# Test the function
calculate_salary(50000, 10000, 5000, 3000, tax=8000, insurance=2000)
```

```python
def student_marks(**marks):
    """Calculate student total and average"""
    total = sum(marks.values())
    avg = total / len(marks) if marks else 0
    
    print("📊 Marks Summary:")
    for subject, score in marks.items():
        print(f"  {subject}: {score}")
    print(f"  Total: {total}")
    print(f"  Average: {avg:.2f}")

# Test
student_marks(math=85, english=90, science=78, history=92)
# Output:
# 📊 Marks Summary:
#   math: 85
#   english: 90
#   science: 78
#   history: 92
#   Total: 345
#   Average: 86.25
```

```python
def employee(**info):
    """Display employee details with formatted salary"""
    salary = info.get('salary', 0)
    if salary:
        info['salary'] = f"Rs. {salary:,.2f}"
    
    print("👤 Employee Details:")
    for key, value in info.items():
        print(f"  {key}: {value}")

# Test
employee(name="Ali", position="Manager", salary=50000, department="IT")
# Output:
# 👤 Employee Details:
#   name: Ali
#   position: Manager
#   salary: Rs. 50,000.00
#   department: IT
```

```python
def shopping_cart(**items):
    """Calculate total with tax (10%)"""
    subtotal = sum(items.values())
    tax = subtotal * 0.10
    total = subtotal + tax
    
    print("🛒 Shopping Cart")
    print("-" * 30)
    for item, price in items.items():
        print(f"  {item}: Rs. {price:.2f}")
    print("-" * 30)
    print(f"  Subtotal: Rs. {subtotal:.2f}")
    print(f"  Tax (10%): Rs. {tax:.2f}")
    print(f"  TOTAL: Rs. {total:.2f}")
    print("-" * 30)
    
    return total

# Test
shopping_cart(laptop=85000, mouse=1500, keyboard=3000, monitor=25000)
```

```python
def outer_greeting(name):
    """Outer function"""
    
    def inner_greeting():
        """Inner function"""
        return f"Hello, {name}!"
    
    # Call inner function
    message = inner_greeting()
    print(message)

# Test
outer_greeting("Ali")
```

```python
def calculator(x, y):
    """Outer function"""
    
    def add():
        return x + y
    
    def multiply():
        return x * y
    
    # Call inner functions
    print(f"Sum: {add()}")
    print(f"Product: {multiply()}")

# Test
calculator(5, 3)
```

```python
def price_with_discount(price):
    """Outer function"""
    
    def discount():
        return price * 0.10  # 10% discount
    
    def final_price():
        return price - discount()
    
    print(f"Original Price: Rs. {price}")
    print(f"Discount (10%): Rs. {discount()}")
    print(f"Final Price: Rs. {final_price()}")

# Test
price_with_discount(1000)
```

```python
def convert_temperature(celsius):
    """Outer function"""
    
    def to_fahrenheit():
        return (celsius * 9/5) + 32
    
    def to_kelvin():
        return celsius + 273.15
    
    print(f"{celsius}°C = {to_fahrenheit():.1f}°F")
    print(f"{celsius}°C = {to_kelvin():.1f}K")

# Test
convert_temperature(25)
```

```python
def power_of(number):
    """Outer function"""
    
    def square():
        return number ** 2
    
    def cube():
        return number ** 3
    
    print(f"Number: {number}")
    print(f"Square: {square()}")
    print(f"Cube: {cube()}")

# Test
power_of(4)
```

```python
def age_info(birth_year):
    """Outer function"""
    current_year = 2026
    
    def calculate_age():
        return current_year - birth_year
    
    def is_adult():
        return calculate_age() >= 18
    
    age = calculate_age()
    adult_status = "Adult" if is_adult() else "Minor"
    
    print(f"Birth Year: {birth_year}")
    print(f"Age: {age} years")
    print(f"Status: {adult_status}")

# Test
age_info(2000)
```
