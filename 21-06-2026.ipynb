```python
employee = {
    101 : {'name': 'khan', 'age' : 22, 'Salary' : 2222222},
    102 : {'name': 'alice', 'age' : 33, 'Salary' : 2222},
    103 : {'name': 'john cena', 'age' : 11, 'Salary' : 22}
}

# Example: Get the salary of employee 101
khan_salary = employee[101]['Salary']
print("Khan's Salary:", khan_salary)  # Output: 2222222

# Example: Get the name of employee 103
name_103 = employee[103]['name']
print("Employee 103 Name:", name_103)  # Output: john cena
```

```python
# Update John Cena's age to 45
employee[103]['age'] = 45

print(employee[103])
# Output: {'name': 'john cena', 'age': 45, 'Salary': 22}
```

```python
# Add a 'city' field to Alice's profile
employee[102]['city'] = 'Karachi'

print(employee[102])
# Output: {'name': 'alice', 'age': 33, 'Salary': 2222, 'city': 'Karachi'}
```

```python
# Add employee 104
employee[104] = {'name': 'sana', 'age': 28, 'Salary': 55000}

print(employee[104])
```

```python
# Loop through and print employee details
for emp_id, info in employee.items():
    print(f"Employee ID: {emp_id}")
    print(f"  Name: {info['name']}")
    print(f"  Age: {info['age']}")
    print(f"  Salary: {info['Salary']}")
    print("-" * 20)
```

```python
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# difference()
print(set_a.difference(set_b))  # Output: {1, 2}

# difference_update()
set_a.difference_update(set_b)
print(set_a)  # Output: {1, 2} (set_a is modified)
```

```python
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# intersection()
print(set_a.intersection(set_b))  # Output: {3, 4}

# intersection_update()
set_a.intersection_update(set_b)
print(set_a)  # Output: {3, 4}
```

```python
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# symmetric_difference()
print(set_a.symmetric_difference(set_b))  # Output: {1, 2, 5, 6}

# symmetric_difference_update()
set_a.symmetric_difference_update(set_b)
print(set_a)  # Output: {1, 2, 5, 6}
```

```python
set_a = {1, 2}
set_b = {3, 4}

# union()
print(set_a.union(set_b))  # Output: {1, 2, 3, 4}

# update()
set_a.update(set_b)
print(set_a)  # Output: {1, 2, 3, 4}
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Method 1: Using intersection()
common_elements = A.intersection(B)
print(common_elements)  # Output: {3, 4}
# Method 2: Using the & operator
common_elements_alt = A & B
print(common_elements_alt)  # Output: {3, 4}
```
